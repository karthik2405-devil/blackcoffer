```python
import requests
from bs4 import BeautifulSoup

def extract_text(url):
  
    headers = {
      "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:60.0) Gecko/20100101 Firefox/60.0"
    } 
    # make a GET request to the webpage
    response = requests.get(url,headers=headers)
    #print(response.content)
    #parse the response HTML using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # extract the title and article text
    title = soup.title.string
   # print(title)
    article=soup.find('article')
    article_text = article.find('div', class_='td-post-content').get_text()
    #print(article_text)
    ans = analyse(article_text)
    return ans
    #print(article_text)
    # save the extracted text to a file with the URL_ID as its filename
    #print(article_text)
```

```python
# import transformers
import nltk
import textstat
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
from nltk.corpus import brown
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize, word_tokenize

# Load the BERT model and tokenizer
# model = transformers.BertModel.from_pretrained('bert-base-uncased')
# tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')
def analyse(text):
  # Tokenize the input text
  # input_ids = tokenizer.encode_plus(text, return_tensors='pt', truncation=True, return_attention_mask=True)
  # Generate the attention mask

  # attention_mask = input_ids["attention_mask"][0]
  # input_ids = input_ids["input_ids"][0]
  sia = SentimentIntensityAnalyzer()
  # Perform sentiment analysis
  # output = model(**input_ids)
  # Get the sentiment score from the output
  # sentiment_score = output[0][:,0]
  score = sia.polarity_scores(text)

# Get the positive score
  positive_score = score["pos"]

  #get the polarity and subjective score from the output
  
  
  #polar = sia.polarity_scores(text)

  blob = TextBlob(text)
  subjectivity = blob.sentiment.subjectivity
  polarity = blob.sentiment.polarity
  #Calculate percentage complex words
  words = word_tokenize(text)

    # Get the number of complex words
  fdist = nltk.FreqDist(w.lower() for w in words if w.isalpha() and nltk.corpus.wordnet.synsets(w))
  num_complex_words = sum(fdist.values())

    # Get the total number of words
  num_words = len(words)

    # Calculate the percentage of complex words
  percentage_complex_words = (num_complex_words / num_words) * 100
  # Calculate other text attributes
  tokens = nltk.word_tokenize(text)
  avg_sentence_length = textstat.avg_sentence_length(text)
  
  fog_index = textstat.gunning_fog(text)

  #calculate avg_words_per_sentence
  sentences = nltk.sent_tokenize(text)
  word_count = sum([len(word_tokenize(sentence)) for sentence in sentences])
    # Get the total number of sentences
  sentence_count = len(sentences)
    # calculate the average words per sentence
  average = word_count/sentence_count

  #calculate complex word count 
  

# Get the number of complex words
  complex_words = [word for word in words if nltk.corpus.wordnet.synsets(word)]
  num_complex_words = len(complex_words)
  

  #
  word_count = len(tokens)
  syllables_per_word = textstat.syllable_count(text) / len(tokens)


  #calculate personal_pronoun
  personal_pronouns = ['I','me','you','he','him','she','her','it','we','us','they','them']
  num_personal_pronouns = sum(word.lower() in personal_pronouns for word in words)

  #calculate avg word length
  char_count = sum(len(word) for word in words)
  word_count = len(words)
  avg_word_length = char_count/word_count
  
  
  # Return the results
  return {
      #'positive_score': sentiment_score,
      'positive_score': score["pos"],
      #'negative_score': 1 - sentiment_score,
      'negative_score': score["neg"],
      #'polarity_score': polar,
      'polarity_score':polarity,
      'subjectivity_score': subjectivity,
      'avg_sentence_length': avg_sentence_length,
      'percentage_complex_words': percentage_complex_words,
      'fog_index': fog_index,
      'avg_words_per_sentence': average,
      'complex_word_count': num_complex_words,
      'word_count': word_count,
      'syllables_per_word': syllables_per_word,
      'personal_pronouns': num_personal_pronouns,
      'avg_word_length': avg_word_length
  }
```
